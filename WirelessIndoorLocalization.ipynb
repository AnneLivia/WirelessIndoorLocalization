# Wireless Indoor Localization

This data set has **2000 rows and 8 columns**, and contains data collected in indoor place by observing signal strengths of seven WiFi signals visible on a smartphone. **The data is used to predict the location in one of the four rooms**. **It was collected to perform experimentation on how wifi signal strengths can be used to determine one of the indoor locations.** 

### Data Dictionary 

**Each attribute is wifi signal strength observed on smartphone.**



                    | Column  | Atrribute   | Definition     	        | Data Type    	| % Null Ratios |
                    |---------|-------------|-----------------------------------------------|-------------- |
                    | 1       | Wifi 1      | Signal strength of wifi 1  | Quantitative 	| 0             |
                    | 2       | Wifi 2      | Signal strength of wifi 2  | Quantitative 	| 0             |
                    | 3       | Wifi 3      | Signal strength of wifi 3  | Quantitative 	| 0             |
                    | 4       | Wifi 4      | Signal strength of wifi 4  | Quantitative 	| 0             |
                    | 5       | Wifi 5      | Signal strength of wifi 5  | Quantitative 	| 0             |
                    | 6       | Wifi 6      | Signal strength of wifi 6  | Quantitative 	| 0             |
                    | 7       | Wifi 7      | Signal strength of wifi 7  | Quantitative 	| 0             |
                    | 8       | Room        | Rooms (1, 2, 3, 4) 	    | Quantitative     | 0             |

# Preparing data for training

```python
from sklearn.model_selection import train_test_split
```

```python
data['room'] = data['room'].map({1 : 0, 2 : 1, 3 : 2, 4 : 3})
```

```python
data['room'].unique()
```

```python
data['room'][:5]
```

```python
data['room'].value_counts()
```

```python
X_full = data.drop('room', axis=1);
y = data['room']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_full, y, test_size=0.2, random_state=120)
```

# Creating pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
dt_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_dt', DecisionTreeClassifier(min_samples_leaf=14, random_state=120))
                       ])
```

```python
ada_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_ada', AdaBoostClassifier(n_estimators=200, learning_rate=0.8, random_state=120))
                       ])
```

```python
rf_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_rf', RandomForestClassifier(n_estimators=200, min_samples_leaf=14, random_state=120))
                       ])
```

```python
ex_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_ex', ExtraTreesClassifier(n_estimators=200, min_samples_leaf=14, random_state=120))
                       ])
```

```python
xgb_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_xbg', XGBClassifier(n_estimators=200, use_label_encoder=False, 
                                                    random_state=120, eval_metric='mlogloss'))
                       ])
```

```python
knn_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_knn', KNeighborsClassifier())
                       ])
```

```python
svm_pipeline = Pipeline(steps=
                       [
                        ('stdscaler', StandardScaler()), 
                        ('model_svm', SVC()) 
                       ])
```

# Multi-layer Perceptron classifier

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
```

```python
# hidden_layer_sizes tuple, length = n_layers - 2 (because there's one input layer and one output layer), default=(100,)
# The ith element represents the number of neurons in the ith hidden layer.
# For architecture 56:25:11:7:5:3:1 with input 56 and 1 output hidden layers will be (25:11:7:5:3). 
# So tuple hidden_layer_sizes = (25,11,7,5,3,).
# In the architecture bellow, there are 3 hidden layers, with 25, 50 and 25 neurons respectivelly
# Activiation: ‘relu’, the rectified linear unit function, returns f(x) = max(0, x)
# solver: ‘adam’ refers to a stochastic gradient-based optimizer proposed by Kingma, Diederik, and Jimmy Ba
# ‘lbfgs’ is an optimizer in the family of quasi-Newton methods.
# If the solver is ‘lbfgs’, the classifier will not use minibatch
# When set to “auto”, batch_size=min(200, n_samples).
# tol=0.0001 (default) is the tolerance of change, n_iter_no_change=10 (default) is the patience

# one hidden layer with 100 neurons
mlp_pipeline_1hlayer = Pipeline(steps=[
    ('norm', MinMaxScaler()),
    ('mlp_1', MLPClassifier(hidden_layer_sizes=(100,), 
                          activation='relu', 
                          solver='adam', 
                          batch_size=64, 
                          random_state=1,
                          verbose=False,
                          early_stopping=True))
])

mlp_pipeline_2hlayers = Pipeline(steps=[
    ('norm', MinMaxScaler()),
    ('mlp_2', MLPClassifier(hidden_layer_sizes=(25, 25), 
                          activation='relu', 
                          solver='adam', 
                          batch_size=64, 
                          random_state=1,
                          verbose=False,
                          early_stopping=True))
])


mlp_pipeline_3hlayers = Pipeline(steps=[
    ('norm', MinMaxScaler()),
    ('mlp_3', MLPClassifier(hidden_layer_sizes=(25, 50, 25), 
                          activation='relu', 
                          solver='adam', 
                          batch_size=64, 
                          random_state=1,
                          verbose=False,
                          early_stopping=True))
])
```

```python
mlps = [mlp_pipeline_1hlayer, mlp_pipeline_2hlayers, mlp_pipeline_3hlayers]
```

```python
# comparing 
best_model_index = -1
best_mean = -1
for i in range(len(mlps)):
    cv_mlp = cross_val_score(mlps[i], X_train, y_train, cv=10);
    mean = cv_mlp.mean()
    if  mean > best_mean:
        best_model_index = i
        best_mean = mean
    print("Model {}: {}".format(mlps[i].named_steps, mean))
```

```python
print("Best model: {}, Score: {}".format(mlps[best_model_index], best_mean))
```

# Training best model

```python
mlps[best_model_index].fit(X_train, y_train)
```

```python
pred_mlp = mlps[best_model_index].predict(X_test)
```

```python
print(classification_report(y_test, pred_mlp))
```

```python
cmmlp = confusion_matrix(y_test, pred_mlp)
ConfusionMatrixDisplay(cmmlp).plot();
```
